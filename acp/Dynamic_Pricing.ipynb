### 2. Exploratory Data Analysis (EDA)

##### 2.1 Data Cleaning

* Deal with Duplicates

```python
df = df.drop_duplicates()
```

* Deal with missing values

```python
null_counts = df.isnull().sum()
print(null_counts)
```

Province has 26424 null values which is less than the 10% of the total data, simply drop.

```python
df = df.dropna(subset=['province'])
```

* Deal with the data that the range doesn't make sense

```python
df.describe(include="all").transpose()
```

```python
df = df[df['current_price'] >= 0]
```

```python
df
```

```python
df.info()
```

* Convert the 'collected_date' column into datetime objects

```python
df['collected_date'] = pd.to_datetime(df['collected_date'])
```

* Create Year, Month and Week Columns

```python
df.loc[:, 'Year'] = df['collected_date'].dt.year
df.loc[:, 'Month'] = df['collected_date'].dt.month
df.loc[:, 'Week'] = df['collected_date'].dt.isocalendar().week
```

* Deal with outliers for column ['current_price']

```python
count_price_over_20 = (df['current_price'] > 20).sum()
print(f"Number of prices over 20: {count_price_over_20}")
```

Price more than 20 only have 578963 records which is less than the 10% of the total data, simply drop the outlier.

```python
# Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['current_price'])
plt.title('Boxplot of Prices')
plt.xlabel('Price')
plt.show()
```

```python
# Histogram and Kernel Density Plot
plt.figure(figsize=(12, 6))
sns.histplot(df['current_price'], kde=True, color='blue', bins=50)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

* Distribution of Current Price per Category

```python
plt.figure(figsize=(12, 8))

sns.boxplot(x='category',
            y='current_price',
            data=df
            )

plt.xticks(rotation=90)

plt.title('Current Price Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Current Price (log scale)')

plt.tight_layout()
plt.show()
```

```python
def month_to_season(month):
    if month in range(3, 5):
        return 'Spring'
    elif month in range(6, 8):
        return 'Summer'
    elif month in range(9, 11):
        return 'Fall'
    else:
        return 'Winter'

df.loc[:, 'Season'] = df['Month'].apply(month_to_season)

plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Season', data=df, hue='is_onsale')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')

plt.xlabel('Season')
plt.ylabel('Count')
plt.title('Count of On Sale vs. Season')

plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4))= plt.subplots(2, 2, figsize=(20, 15))
                                            
sns.lineplot(x = 'Year', y = 'current_price', data=df, ax=ax1)
sns.lineplot(x = 'Month', y = 'current_price', data=df, ax=ax2)
sns.lineplot(x = 'Week', y = 'current_price', data=df, ax=ax3)
sns.lineplot(x = 'Season', y = 'current_price', data=df, ax=ax4)

ax1.set_title('Yearly Distribution')
ax2.set_title('Monthly Distribution')
ax3.set_title('Weekly Distribution')
ax4.set_title('Season Distribution')

plt.tight_layout()
plt.show()
```
