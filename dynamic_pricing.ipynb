CREATING YEAR AND WEEK COLUMNS

```python
df['data_year_week'] = df['data_year_week'].astype(str)

df['Year'] = df['data_year_week'].str[:4].astype(int)
df['Week'] = df['data_year_week'].str[4:].astype(int)
```

CONVERTING KILOGRAMS TO GRAMS

```python
df['standard_size_value'] = df.apply(lambda row: row['standard_size_value'] * 1000 if row['standard_size_unit'] == 'kilogram' else row['standard_size_value'], axis=1)
df['standard_size_unit'] = df['standard_size_unit'].replace({'kilogram': 'g'})

df.head()
```

```python
# Function to map week to season
def week_to_season(week):
    if week in range(1, 14):
        return 'Spring'
    elif week in range(14, 27):
        return 'Summer'
    elif week in range(27, 40):
        return 'Fall'
    else:
        return 'Winter'

# Apply the function to create a new 'Season' column
df['Season'] = df['Week'].apply(week_to_season)

# Set up the count plot using seaborn
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Season', data=df, hue='onPromo')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')


# Add labels and title
plt.xlabel('Season')
plt.ylabel('Count')
plt.title('Count of On Promo vs. Season')

# Show the plot
plt.show()
```
