```python
df['standard_size_value'] = df.apply(lambda row: row['standard_size_value'] * 1000 if row['standard_size_unit'] == 'kilogram' else row['standard_size_value'], axis=1)
df['standard_size_unit'] = df['standard_size_unit'].replace({'kilogram': 'g'})

df.head()
```

```python

# Use the sum() function to calculate the count of True values
true_count = df['onPromo'].sum()

print(f"The count of True values is: {true_count}")
```

```python
# Function to map week to season
def week_to_season(week):
    if week in range(1, 14):
        return 'Spring'
    elif week in range(14, 27):
        return 'Summer'
    elif week in range(27, 40):
        return 'Fall'
    else:
        return 'Winter'

# Apply the function to create a new 'Season' column
df['Season'] = df['Week'].apply(week_to_season)

# Set up the count plot using seaborn
plt.figure(figsize=(12, 6))
ax = sns.countplot(x='Season', data=df, hue='onPromo')

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='baseline')


# Add labels and title
plt.xlabel('Season')
plt.ylabel('Count')
plt.title('Count of On Promo vs. Season')

# Show the plot
plt.show()
```

```python
#Histogram and Kernel Density Plot
plt.figure(figsize=(12, 6))
sns.histplot(df['price'], kde=True, color='blue', bins=50)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['price'], color='green')
plt.title('Boxplot of Prices')
plt.xlabel('Price')
plt.show()
```

```python
#Kernel Density Estimation
plt.figure(figsize=(12, 6))
sns.kdeplot(df['price'], color='purple', fill=True)
plt.title('Kernel Density Estimation of Prices')
plt.xlabel('Price')
plt.ylabel('Density')
plt.show()
```

```python
#Set Price Floors and Ceilings
# Assuming your business rule is prices between 10 and 100
min_price = 10
max_price = 100
valid_prices = df[(df['price'] >= min_price) & (df['price'] <= max_price) & (df['onPromo'] == True)]
```

```python
#Compare Discount Rates
df['discount_rate'] = ((df['regular_price_ea'] - df['current_price_ea']) / df['regular_price_ea']) * 100
discounted_prices = df[df['onPromo'] == True]

plt.figure(figsize=(12, 6))
sns.boxplot(x=discounted_prices['discount_rate'], color='orange')
plt.title('Boxplot of Discount Rates for Discounted Items')
plt.xlabel('Discount Rate')
plt.show()
```
