```python
resource_id = 'c93fdac9-a806-440e-bac9-0aad58765227'
limit = 1500
offset = 0
api_url = 'https://data.ontario.ca/api/3/action/datastore_search?resource_id={}&limit={}&offset={}'.format(resource_id,limit,offset)

print(api_url)
```

This dataset presents a comprehensive summary of research projects that have received funding from the Ministry of Colleges and Universities. 

We need to provide resource id, limit, offset and api's url. Also, we are using inherit auth from parent postman. By doing so, by default, requests inside the collection or folder will inherit auth from the parent.

```python
import pandas as pd
import sqlalchemy as sa
import requests #necessary to get request from API
from sqlalchemy import Column
```

```python
api_response = requests.get(api_url)
```

```python
api_response
```

```python
data = api_response.json()
# Parse the JSON response into a Python data structure
data
```

```python
data['result']['records']
```

```python
Research_Funding_cases = pd.DataFrame(data['result']['records'])
Research_Funding_cases
```

### Data Engineering

```python
Research_Funding_cases['approval_date'] = pd.to_datetime(Research_Funding_cases['Approval Date'])
Research_Funding_cases['Time_Lag'] = Research_Funding_cases['approval_date'].diff(1)#difference with the previous record

relevant_columns = ['approval_date','Time_Lag']
Research_Funding_cases
```

```python
Research_Funding_cases['Time_Lag'] = Research_Funding_cases['Time_Lag'].astype(str)
```

```python
clean_copy_df = Research_Funding_cases[relevant_columns]
```
